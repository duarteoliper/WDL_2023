```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load CSV file into NumPy array

path = os.path.join(os.pardir, 'reports', 'iteration_results.csv')
path
```

```python
import re
data = pd.read_csv(path, delimiter=',', names = ['timestamp', 'window', 'train_loss', 'val_loss'])
data['train_loss'] = data['train_loss'].apply(lambda x: x[1:-1])
data['val_loss'] = data['val_loss'].apply(lambda x: x[1:-1])
data = data.astype({'window': int, 'train_loss': float, 'val_loss': float})
```

```python
data.plot('window', ['train_loss', 'val_loss'])
```
